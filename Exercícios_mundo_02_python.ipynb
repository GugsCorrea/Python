```python
#desafio 70 - sacola de compras

print('-'*30)
print('     CARRINHO DE COMPRAS')
print('-'*30)

#variaveis
produto = str(input('Nome do Produto: ')).title()
preco = float(input('Valor do Produto: R$'))
total = caros = 0
barato = preco
nome = produto

while True:
  if preco < barato:
    barato = preco
    nome = produto
  if preco > 1000:
    caros += 1
  cont = str(input('Continuar comprando [s/n]? ')).lower()
  total += preco
  if cont == 's':
    produto = str(input('''
Nome do Produto: ''')).title()
    preco = float(input('Valor do Produto: R$'))
  if cont == 'n':
    break

print(f'''

Total do carrinho: R${total}''')
print(f'Produtos acima de R$1000.00: {caros}')
print(f'''Produto mais barato: {nome}
Valor: {preco}''')
```
